```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
dataset_path = "aalesund_2024_10_10_11_53_38 (1).csv"
df = pd.read_csv(dataset_path)
```

```python
competitors = ['bauhaus',
       'byggmakker', 'byggmax', 'byggtorget', 'elektroimportoren', 'fargerike', 
     'gausdal_landhandleri', 'industrivarer', 'jemogfix',
       'jernia', 'maxbo', 'megaflis', 'monter', 'obs_bygg', 'staypro', 'torn',
       'verktoy', 'xl_bygg']

df["Row_mean"] = df[competitors].mean(axis=1)
df["CPI"] = df["byggern"]/df["Row_mean"]
df.head()
```

```python
df = df[['Kode', 'Navn', 'Enhet', 'Kommentar', 'bauhaus', 'byggern',
       'byggmakker', 'byggmax', 'byggtorget', 'elektroimportoren', 'fargerike',
    'gausdal_landhandleri', 'industrivarer', 'jemogfix',
       'jernia', 'maxbo', 'megaflis', 'monter', 'obs_bygg', 'staypro', 'torn',
       'verktoy', 'xl_bygg', 'Row_mean', 'CPI']]
```

```python
df = df[df["byggern"].notna()]
df.describe()
```

```python
fig = px.scatter(df, x="CPI",title="Simple Scatter Plot")
fig.show()
```

```python
min_row = df.loc[df['CPI'].idxmin()]
low_CPI = df[df["CPI"] < 0.7]
df  = df[df["CPI"] >= 0.3]
df = df[df["CPI"] <= 1.9]
df.describe()
```

```python
fig = px.scatter(df, x="CPI",title="Simple Scatter Plot")
fig.show()
```
