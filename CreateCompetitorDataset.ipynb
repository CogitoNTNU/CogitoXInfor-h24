```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from random import randint

filepath = 'ECOMMRecords2020.csv'
data = pd.read_csv(filepath)
data.head()
```

```python
df_comps = data.groupby("ProductID")["Sales"].mean().reset_index()
df_comps = df_comps.rename(columns={"Sales": "byggern"})

df_comps.head()
```

```python
def fill_dataset(df, competitors, variation_percentage):
    result_df = df.copy()
    
    for competitor in competitors:
        # Create a new column for each competitor
        result_df[competitor] = result_df['byggern'].apply(
            lambda x: x * (1 + np.random.uniform(-variation_percentage/100, variation_percentage/100))
        )
    
    return result_df
```

```python
competitors = ["bauhaus", "byggmakker", "byggmax", "byggtorget"]
df_comps = fill_dataset(df_comps, competitors, 20)

df_comps
```
